```python
from nltk.corpus import PlaintextCorpusReader
import sys
wordlists = PlaintextCorpusReader("Nepali_Corpus", '.*txt')
```

```python
texts,invalid = [],0
for i in wordlists.fileids():
    try:
        texts.append(wordlists.raw(i).replace("\ufeff",""))
    except UnicodeDecodeError:
        invalid+=1
```

```python
len(wordlists.fileids())
```

```python
print(len(texts),invalid)
```

```python
import re
##Split in text
paragraph = list()
for text in texts:
    raw = re.split(r"\n|\r",text.strip())
    if len(text) > 0 or not re.match('(\s)+',text): 
        paragraph.extend(raw)
print(len(paragraph))
```

```python
#Split Into Sentences
pura_sents = []
for i in paragraph:
    pura_sents.extend(re.split(r'(?<=।)\s',i))
print(len(pura_sents))
```

```python
purna_prasna_sents = []
for i in pura_sents:
    purna_prasna_sents.extend(re.split(r'(?<=\?)\s',i))
#sents
print(len(purna_prasna_sents))
```

```python
sents = []
for i in purna_prasna_sents:
    sents.extend(re.split(r'(?<=\!)\s',i))
#sents
print(len(sents))
```

```python
non_dev_nagari = []
for i in sents:
    invalid = False
    for j in i:
        #Devanagari range 
        if not 0x0090 <= ord(j) <= 0x97F: 
            non_dev_nagari.append(j)
```

```python
remove_candidates = set(non_dev_nagari)
```

```python
#\u200d is ZWJ
#\xa0 is non breaking space
remove_candidates-={" ","!","(",")","–","’","‘",",",".","?","-","/",":","“","”","\u200d","\xa0","'"}

remove_candidates
```

```python
invalid_char_sents = []
chr_valid_sents = []
for i in sents:
    invalid = False
    invalid_chars = set()
    for j in i:
        #Devanagari range 
        if j in remove_candidates: 
            invalid = True
            invalid_chars.update((j))
    if invalid :
        invalid_char_sents.append(i+"**"+"*".join(invalid_chars)) #For filtering
    else:
        chr_valid_sents.append(i)
len(invalid_char_sents)
```

```python
len(chr_valid_sents)
```

```python
##Remove Puntuation..#I know I should use regex...But it was easir to do it this way :D
#(",")","–","’","‘",",",".","?","-","/",":","“","”","\u200d","\xa0","'"
punct_removed_sents = []
for i in chr_valid_sents:
    punct_removed_sents.append(i.replace("'","")
                              .replace(','," ")
                                .replace('’',"")
                               .replace('‘',"")
                               .replace('–'," ")
                               .replace('\xa0'," ")#Space Character 
                               .replace("'","")
                               .replace("“","")
                               .replace("”","")
                               .replace("-","")
                               .replace("?","")
                               .replace("।","")
                               .replace("!","")
                              )
punct_removed_sents
```

```python
min_valid_sent_lenght = 5
valid_full_len_sents = list(filter(lambda x: len(x) > min_valid_sent_lenght,
                            punct_removed_sents ))
print(len(valid_full_len_sents))
```

```python
augumented_sents = []
for i in valid_full_len_sents:
    augumented_sents.append("BoS " + i + " EoS")
```

```python
import pickle
with open('sents.pkl','rb') as pk:
    trainning_data = pickle.load(pk)
```

```python
##MY CPU(NO GPU<SAD>) WILL TAKE INF TIME TO HANDLE THE WHOLE DATASET
trainning_data = trainning_data[:100]
```

```python
import random
import copy
from keras.preprocessing.sequence import pad_sequences
from keras.utils import  to_categorical
def gen_batch(X,Y,batch_size):
    while True:    
        random.shuffle(X)
        random.shuffle(Y)
        start = 0
        while start + batch_size < len(X):
            yield pad_sequences(X[start:start + batch_size],maxlen=max_sent_len),to_categorical(Y[start:start + batch_size],num_classes=relevant_word)
            start+=batch_size
max_sent_len
```

```python
batch_size = 128
```

```python
from sklearn.model_selection import train_test_split
```

```python
train,test = train_test_split(trainning_data,test_size=0.2)
```

```python
X_train = []
Y_train = []
for sent in train:
    for i in range(1,len(sent)-1):
        X_train.append(sent[:i])
        Y_train.append(sent[i])
```

```python
X_test = []
Y_test = []
for sent in test:
    for i in range(1,len(sent)-1):
        X_test.append(sent[:i])
        Y_test.append(sent[i])
```

```python
from keras.preprocessing.sequence import pad_sequences
```

```python
###Now Buliding Model
```

```python
from keras.layers import Dense, LSTM,Dropout,Embedding
from keras.models import Sequential
```

```python
model = Sequential()
```

```python
model.add(Embedding(relevant_word+1,100,input_length=max_sent_len))
model.add(Dropout(0.2))
```

```python
model.add(LSTM(128))
model.add(Dropout(0.2))
```

```python
model.add(Dense(relevant_word))
```

```python
model.compile(optimizer='Adam',loss='categorical_crossentropy')
```

```python
model.summary()
```

```python
total_epoch_train = len(X_train)//batch_size
total_epoch_val = len(X_test)//batch_size
total_epoch_val
```

```python
model.fit_generator(gen_batch(X_train,Y_train,batch_size),
                        steps_per_epoch=total_epoch_train,
                        epochs=20,
                    validation_data = gen_batch(X_test,Y_test,batch_size),
                        validation_steps =total_epoch_val)
```
