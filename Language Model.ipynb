```python
from nltk.corpus import PlaintextCorpusReader
import sys
wordlists = PlaintextCorpusReader("Nepali_Corpus", '.*txt')
```

```python
texts,invalid = [],0
for i in wordlists.fileids():
    try:
        texts.append(wordlists.raw(i).replace("\ufeff",""))
    except UnicodeDecodeError:
        invalid+=1
```

```python
len(wordlists.fileids())
```

```python
print(len(texts),invalid)
```

```python
import re
##Split in text
paragraph = list()
for text in texts:
    raw = re.split(r"\n|\r",text.strip())
    if len(text) > 0 or not re.match('(\s)+',text): 
        paragraph.extend(raw)
print(len(paragraph))
```

```python
#Split Into Sentences
pura_sents = []
for i in paragraph:
    pura_sents.extend(re.split(r'(?<=।)\s',i))
print(len(pura_sents))
```

```python
purna_prasna_sents = []
for i in pura_sents:
    purna_prasna_sents.extend(re.split(r'(?<=\?)\s',i))
#sents
print(len(purna_prasna_sents))
```

```python
sents = []
for i in purna_prasna_sents:
    sents.extend(re.split(r'(?<=\!)\s',i))
#sents
print(len(sents))
```

```python
non_dev_nagari = []
for i in sents:
    invalid = False
    for j in i:
        #Devanagari range 
        if not 0x0090 <= ord(j) <= 0x97F: 
            non_dev_nagari.append(j)
```

```python
remove_candidates = set(non_dev_nagari)
```

```python
#\u200d is ZWJ
#\xa0 is non breaking space
remove_candidates-={" ","!","(",")","–","’","‘",",",".","?","-","/",":","“","”","\u200d","\xa0","'"}

remove_candidates
```

```python
invalid_char_sents = []
chr_valid_sents = []
for i in sents:
    invalid = False
    invalid_chars = set()
    for j in i:
        #Devanagari range 
        if j in remove_candidates: 
            invalid = True
            invalid_chars.update((j))
    if invalid :
        invalid_char_sents.append(i+"**"+"*".join(invalid_chars)) #For filtering
    else:
        chr_valid_sents.append(i)
len(invalid_char_sents)
```

```python
len(chr_valid_sents)
```

```python
##Remove Puntuation..#I know I should use regex...But it was easir to do it this way :D
#(",")","–","’","‘",",",".","?","-","/",":","“","”","\u200d","\xa0","'"
punct_removed_sents = []
for i in chr_valid_sents:
    punct_removed_sents.append(i.replace("'","")
                              .replace(','," ")
                                .replace('’',"")
                               .replace('‘',"")
                               .replace('–'," ")
                               .replace('\xa0'," ")#Invalid Unicode Character
                               .replace("'","")
                               .replace("“","")
                               .replace("”","")
                               .replace("-","")
                               .replace("?","")
                               .replace("।","")
                               .replace("!","")
                              )
punct_removed_sents
```

```python
min_valid_sent_lenght = 5
valid_full_len_sents = list(filter(lambda x: len(x) > min_valid_sent_lenght,
                            punct_removed_sents ))
print(len(valid_full_len_sents))
```

```python
valid_full_len_sents
```

```python
trainning_data = t.texts_to_sequences(valid_full_len_sents)
trainning_data
```
